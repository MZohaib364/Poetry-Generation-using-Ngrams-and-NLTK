```python
import string

starting_words = []
ending_words = []

lines = text.split('\n')

# For storing starting and ending word of each line
for line in lines:
  temp_tokens = nltk.word_tokenize(line)
  if len(temp_tokens)>0:
    starting_words.append(temp_tokens[0])
    ending_words.append(temp_tokens[-1])

# Tokenizing original text and removing punctuations
unfiltered_tokens = nltk.word_tokenize(text)

punctuation = string.punctuation
tokens = []
for t in unfiltered_tokens:
  if t not in punctuation:
    tokens.append(t)
```

```python
unigrams = list(nltk.ngrams(tokens, 1))
unigramFD = nltk.FreqDist(unigrams)

bigrams = list(nltk.ngrams(tokens, 2))
bigramFD = nltk.FreqDist(bigrams)

trigrams = list(nltk.ngrams(tokens, 3))
trigramFD = nltk.FreqDist(trigrams)
```
